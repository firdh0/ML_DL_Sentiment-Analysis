# ðŸŽ¯ **Step 2: Split data into feature and label**

```python
X = df_tweet['clean_tweet_sentence']

y = df_tweet['sentimen']
```

# ðŸŽ¯ **Step 3: Processing X**
---

## âœ¨ 3.1 - Create Char Level Corpus for Tokenizing and Padding

```python
X = [str(text) for text in X]

tokenizer_char_level = Tokenizer(num_words=None, oov_token='<OOV>', char_level=True)
tokenizer_char_level.fit_on_texts(X)
```

```python
len(tokenizer_char_level.word_index)
```

```python
first_10_items = {k: tokenizer_char_level.word_index[k] for k in list(tokenizer_char_level.word_index)[:10]}
print(first_10_items)
```

```python
tokenizer_char_json = tokenizer_char_level.to_json()

# Menyimpan representasi JSON ke dalam sebuah file
with open('/content/drive/MyDrive/INDONESIA AI/Sentiment Analysis/Tokenizer/tokenizer_char_level.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(tokenizer_char_json))
```

## âœ¨ 3.1 - Create Word Level Corpus for Tokenizing and Padding

```python
tokenizer_word_level = Tokenizer(num_words=None, oov_token='<OOV>')
tokenizer_word_level.fit_on_texts(X)
```

```python
len(tokenizer_word_level.word_index)
```

```python
first_10_items = {k: tokenizer_word_level.word_index[k] for k in list(tokenizer_word_level.word_index)[:10]}
print(first_10_items)
```

```python
tokenizer_word_json = tokenizer_word_level.to_json()

# Menyimpan representasi JSON ke dalam sebuah file
with open('/content/drive/MyDrive/INDONESIA AI/Sentiment Analysis/Tokenizer/tokenizer_word_level.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(tokenizer_word_json))
```
